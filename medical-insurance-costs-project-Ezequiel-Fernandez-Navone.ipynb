```python
#The first class I am making, is the one to get averages from a certain dictionary or list: 

class Average():
    total_people = len(dict_file)
    def get_average(self, column, number):
        result = Decimal(number / self.total_people)
        output = round(result,2)
        print("The average {col} for a person in this dataset is {num}.".format(col=column, num=output))
        
    def get_average_in_list(self, one_list, column):
        number = 0
        for row in one_list:
            number += float(row)
        result = Decimal(number / self.total_people)
        output = round(result,2)
        print("The average {col} for a person in this dataset is {num}.".format(col=column, num=output))
        

#My test to see the results of calling this class:

average_children = Average()
average_children.get_average("children", children)

average_age = Average()
average_age.get_average_in_list(ages, "age")

average_bmi = Average()
average_bmi.get_average_in_list(bmis, "bmi")

average_charges = Average()
average_charges.get_average_in_list(charges, "charges")

```

```python
#Now, a class to get the highest and lowest age:

class Limits():
    transition_number_min = 1000000000000
    transition_number_max = 0
    def get_min_and_max(self,one_list,column):
        for row in one_list:
            if int(row) < self.transition_number_min:
                self.transition_number_min = int(row)
        print("The minimum {col} in this dataset is {num}.".format(col=column, num=self.transition_number_min))
        for row in one_list:
            if int(row) > self.transition_number_max:
                self.transition_number_max = int(row)
        print("The maximum {col} in this dataset is {num}.".format(col=column, num=self.transition_number_max))


#I am calling it here:

age_min = Limits()
age_min.get_min_and_max(ages, "age")
```

```python
#I am printing my previous gender variables to see how many male and female entries are in this dataset:

print("Male people in this dataset = " + str(is_male))
print("Female people in this dataset = " + str(is_female))

#Now, I am defining two functions to know the average of a certain column, working with separate genders:

def average_charges_by_gender(one_list, gender):
    addition = 0
    for row in one_list:
        addition += float(row['charges'])
        result = (addition/len(one_list))
        output = round(result,2)
    print("The average charges for a {gen} person in this dataset is ${res}.".format(gen=gender,res=str(output)))
    
def average_age_by_gender(one_list, gender):
    addition = 0
    for row in one_list:
        addition += float(row['age'])
        result = (addition/len(one_list))
        output = round(result,2)
    print("The average age for a {gen} person in this dataset is {res}.".format(gen=gender,res=str(output)))


#My call to see the output of my functions:

average_charges_by_gender(male_list,"male")
average_charges_by_gender(female_list, "female")

average_age_by_gender(male_list, "male")
average_age_by_gender(female_list, "female")
```

```python
#I am making another two classes, to work on the bmi data I am given:

class BmiInformation():
    bmi_min = 1000000000000
    bmi_max = 0
    total_people = len(bmis)
    def get_min_and_max(self,one_list):
        for row in one_list:
            if float(row) < self.bmi_min:
                self.bmi_min = float(row)
        print("The minimum {col} in this dataset is {num}.".format(col="bmi", num=self.bmi_min))
        for row in one_list:
            if float(row) > self.bmi_max:
                self.bmi_max = float(row)
        print("The maximum {col} in this dataset is {num}.".format(col="bmi", num=self.bmi_max))
    def get_average(self, one_list):
        total_bmi = 0
        for row in one_list:
            total_bmi += float(row)
        result = Decimal(total_bmi / self.total_people)
        output = round(result,2)
        print("The average {col} for a person in this dataset is {num}.".format(col="bmi", num=output))
        
class BmiComparison():
    def comparison(self,one_list):
        three_bmis_list = []
        for row in one_list:
            if row['bmi'] == '15.96':
                three_bmis_list.append(row)
            if row['bmi'] == '53.13':
                three_bmis_list.append(row)
            if row['bmi'] == '30.5':
                three_bmis_list.append(row)
        for row in three_bmis_list:
            print("For a bmi of {a}, the charge is ${b}.".format(a=row['bmi'], b=row['charges']))
    
        
#I am calling these classes here:
        
get_bmi = BmiInformation()
get_bmi.get_min_and_max(bmis)
get_bmi.get_average(bmis)
compare_bmis = BmiComparison()
compare_bmis.comparison(dict_file)
```

```python
#I proceeded to work with two more factors in this dataset, which I think can give me really insightful data.
#I defined four functions below:

def smoker_impact(one_list):
    sum_charges = 0
    for row in one_list:
        sum_charges += float(row['charges'])
    average = round(sum_charges / len(one_list),2)
    print("The average charge for a smoker in this dataset is ${a}.".format(a=average))
    
def non_smoker_impact(one_list):
    sum_charges = 0
    for row in one_list:
        sum_charges += float(row['charges'])
    average = round(sum_charges / len(one_list),2)
    print("The average charge for a non smoker in this dataset is ${a}.".format(a=average))
    
def kids_impact(one_list):
    sum_charges = 0
    for row in one_list:
        sum_charges += float(row['charges'])
    average = round(sum_charges / len(one_list),2)
    print("The average charge for someone with children in this dataset is ${a}.".format(a=average))
    
def no_kids_impact(one_list):
    sum_charges = 0
    for row in one_list:
        sum_charges += float(row['charges'])
    average = round(sum_charges / len(one_list),2)
    print("The average charge for someone with no children in this dataset is ${a}.".format(a=average))  


#Here I am calling the functions after I printed how many smokers are on this dataset:

print("There are {yes} people who smoke and {no} who does not smoke.".format(yes=is_smoker, no=is_not_smoker))
smoker_impact(smoker_list)
non_smoker_impact(not_smoker_list)
kids_impact(has_children_list)
no_kids_impact(has_no_children_list)
```

```python
#This class is made to see if there there is a big difference of age from people with no children, or ones who has one or more:

class AveragePerChild():
    def get_average(self, one_list, num_of_kids):
        total_people = len(one_list)
        number = 0
        for row in one_list:
            number += int(row['age'])
        result = Decimal(number / total_people)
        output = round(result,2)
        if num_of_kids == "0":
            print("The average age for a person with no children in this dataset is {num}.".format(num=output))
        if num_of_kids == "1":
            print("The average age for a person with one child in this dataset is {num}.".format(num=output))
        if num_of_kids == "3":
            print("The average age for a person with three children in this dataset is {num}.".format(num=output))
            
get_average_per_child = AveragePerChild()
get_average_per_child.get_average(has_no_children_list, "0")
get_average_per_child.get_average(has_1_child_list, "1")
get_average_per_child.get_average(has_3_children_list, "3")
```

```python
#I am studying the region factor, first printing how many people are from each region:

print("There are " + str(from_southeast) + " people from southeast.")
print("There are " + str(from_southwest) + " people from southwest.")
print("There are " + str(from_northwest) + " people from northwest.")
print("There are " + str(from_northeast) + " people from northeast.")

#Now, I am creating another class, to see the average charges per region:

class Average_per_region():
    def get_averages(self,one_dict,region):
        charges = 0
        length = 0
        for row in one_dict:
            if row['region'] == region:
                charges += float(row['charges'])
                length += 1
        average = charges/length
        result = round(average,2)
        print("The average charge for the {reg} region is ${res}.".format(reg=region, res = result))

#I am calling the function under this class to see the output:
        
averages = Average_per_region()
averages.get_averages(dict_file, "southeast")
averages.get_averages(dict_file, "southwest")
averages.get_averages(dict_file, "northwest")
averages.get_averages(dict_file, "northeast")
```

```python
#The difference between the least and highest charge in this dataset has to show me difference in key factors.
#In order to see this, I am defining one another class:

class ChargesInformation(): 
    charge_min = 1000000000000
    charge_max = 0
    max_and_min = []
    def get_min_and_max_charge(self,one_list):
        for row in one_list:
            if float(row['charges']) < self.charge_min:
                self.charge_min = float(row['charges'])
        print("The minimum charge in this dataset is {num}.".format(num=self.charge_min))
        for row in one_list:
            if float(row['charges']) > self.charge_max:
                self.charge_max = float(row['charges'])
        print("The maximum charge in this dataset is {num}.".format(num=self.charge_max))
        for row in one_list:
            if float(row['charges']) == self.charge_min:
                self.max_and_min.append(row)
            if float(row['charges']) == self.charge_max:
                self.max_and_min.append(row)
        print("This is what I know about the least and the most charged person in this dataset:")
        print(self.max_and_min)
        

#Now, I can see the two entries and compare each other:

charges_limits = ChargesInformation()
charges_limits.get_min_and_max_charge(dict_file)
```

```python
#Finally, I think age has to be a significant factor. With a final class created, I aim to take a look at that:

class ChargeByAge():
    charges_of_20 = []
    charges_of_30 = []
    charges_of_40 = []
    charges_of_50 = []
    charges_of_60 = []
    
    
    def get_charges(self,one_list):
        for row in one_list:
            if int(row['age']) >= 20 and int(row['age']) < 30:
                self.charges_of_20.append(row['charges'])
            if int(row['age']) >= 30 and int(row['age']) < 40:
                self.charges_of_30.append(row['charges'])
            if int(row['age']) >= 40 and int(row['age']) < 50:
                self.charges_of_40.append(row['charges'])
            if int(row['age']) >= 50 and int(row['age']) < 60:
                self.charges_of_50.append(row['charges'])
            if int(row['age']) >= 60 and int(row['age']) < 70:
                self.charges_of_60.append(row['charges'])
        number_20 = 0
        number_30 = 0
        number_40 = 0
        number_50 = 0
        number_60 = 0
        for row in self.charges_of_20:
            number_20 += float(row)
        result = Decimal(number_20 / len(self.charges_of_20))
        output = round(result,2)
        print("The average charge for a person bewteen 20 and 29 is {num}.".format(num=output))
        for row in self.charges_of_30:
            number_30 += float(row)
        result = Decimal(number_30 / len(self.charges_of_30))
        output = round(result,2)
        print("The average charge for a person bewteen 30 and 39 is {num}.".format(num=output))
        for row in self.charges_of_40:
            number_40 += float(row)
        result = Decimal(number_40 / len(self.charges_of_40))
        output = round(result,2)
        print("The average charge for a person bewteen 40 and 49 is {num}.".format(num=output))
        for row in self.charges_of_50:
            number_50 += float(row)
        result = Decimal(number_50 / len(self.charges_of_50))
        output = round(result,2)
        print("The average charge for a person bewteen 50 and 59 is {num}.".format(num=output))
        for row in self.charges_of_60:
            number_60 += float(row)
        result = Decimal(number_60 / len(self.charges_of_60))
        output = round(result,2)
        print("The average charge for a person bewteen 60 and 69 is {num}.".format(num=output))
                                                    
        
#The class has a lot of code, but I can call it in a really simple way:

average_by_age = ChargeByAge()
average_by_age.get_charges(dict_file)
```
